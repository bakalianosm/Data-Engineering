```python
import matplotlib.pyplot as plt
import string
```

```python
def extract_letter_occurrences(file_path):
    word_counts = {}
    with open(file_path, 'r') as file:
        for line in file:
            word, count = line.strip().split('\t')  # Split the line into letter and count
            word_counts[word] = int(count)  # Convert count to integer
    return word_counts

# Function to plot letter occurrences
def plot_letter_occurrences(word_counts):
    letters = list(word_counts.keys())
    counts = list(word_counts.values())
    plt.figure(figsize=(10, 6))
    plt.bar(letters, counts)
    plt.xlabel('Letter')
    plt.ylabel('Counts')
    plt.title('Counts for Each Word', pad = 40)
    for i, v in enumerate(word_counts.values()):
        plt.text(i, v + 0.1 * max(word_counts.values()), str(v), ha='center', va='bottom')
    plt.show()

# Example usage
file_path = 'part-r-00000'  # Replace 'your_file.txt' with the path to your file
word_counts = extract_letter_occurrences(file_path)
plot_letter_occurrences(word_counts)
```
